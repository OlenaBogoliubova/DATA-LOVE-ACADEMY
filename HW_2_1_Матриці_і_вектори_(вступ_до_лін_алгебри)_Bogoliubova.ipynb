```python
import numpy as np

# Вектор — варіант 1: random.uniform
import random
vector_uniform = np.array([random.uniform(0, 1) for _ in range(10)])

print("Вектор (random.uniform):")
print(vector_uniform)
print("Розмір:", vector_uniform.shape)
```

```python
# Вектор — варіант 2: np.random.rand
vector = np.random.rand(10)

print("\nВектор (np.random.rand):")
print(vector)
print("Розмір:", vector.shape)
```

```python
# Матриця 3x3
matrix = np.array([
    [11, 14, 17],
    [20, 23, 26],
    [29, 32, 35]
])

print("\nМатриця:")
print(matrix)
print("Розмір:", matrix.shape)
```
