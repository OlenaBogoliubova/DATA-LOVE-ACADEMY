1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return 10*x - 5

xs = np.linspace(0,10,20)
plt.plot(xs, f(xs), '*');
```

```python
import sympy as sp

x = sp.symbols('x')
f_lin = 10*x - 5
derivative_f_lin = sp.diff(f_lin, x)
```

```python
derivative_f_lin
```

```python
x = np.linspace(0, 5, 100)
f = 10*x - 5
df = np.full_like(x, 10)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10*x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')
```

Похідна в нас константа (10). Це означає, що швидкість зміни оцінки є постійною при будь-якій кількості годин.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Derivative sin(x):", derivative_f_sin)
print("Derivative ln(x):", derivative_f_ln)
```

```python
x_np = np.linspace(0, 5, 100)

x_sym = sp.symbols('x_sym')

f_sym = sp.sin(x_sym)
df_sym = sp.diff(f_sym, x_sym)

# Lambdify the symbolic functions to create numerical functions
f_numeric = sp.lambdify(x_sym, f_sym, 'numpy')
df_numeric = sp.lambdify(x_sym, df_sym, 'numpy')

# Calculate numerical values for plotting
f = f_numeric(x_np)
df = df_numeric(x_np)

plt.figure(figsize=(10, 5))
plt.plot(x_np, f, label='f(x) = sin(x)', color='blue')
plt.plot(x_np, df, label="f'(x) = cos(x)", color='green')

plt.title("Function vs Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x_np = np.linspace(0, 5, 100)

x_sym = sp.symbols('x_sym')

f_sym = sp.ln(x_sym)
df_sym = sp.diff(f_sym, x_sym)

# Lambdify the symbolic functions to create numerical functions
f_numeric = sp.lambdify(x_sym, f_sym, 'numpy')
df_numeric = sp.lambdify(x_sym, df_sym, 'numpy')

# Calculate numerical values for plotting
f = f_numeric(x_np)
df = df_numeric(x_np)

plt.figure(figsize=(10, 5))
plt.plot(x_np, f, label='f(x) = ln(x)', color='blue')
plt.plot(x_np, df, label="f'(x) = 1/x", color='green')

plt.title("Function vs Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
